```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_json('/Users/ryan/text-query-tweets.json', lines=True)
```

```python
elephant_df = df.loc[df['renderedContent'].str.contains('elephant') == True]
```

```python
elephant_tweets = []
for tweet in elephant_df['renderedContent']:
    elephant_tweets.append(tweet)
```

```python
for tweet in elephant_tweets:
    print(tweet)
    print()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
labels = ['Contains Elephants', "Doesn't contain elephants"]
count_tweets = [len(elephant_tweets), (500 - len(elephant_tweets))]
ax.bar(labels,count_tweets)
plt.show()
```
